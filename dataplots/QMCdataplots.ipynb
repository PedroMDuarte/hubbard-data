## First gather up all of Thereza's data in a folder called QMC_Final.   Since not the same data is available for entropy, density, and Spi I will treat them separately:


* ### Data that has Spi is named TX.XXX.dat

* ###  Data that has entropy is named SX.XXX.dat  

* ### Data that has density is named nX.XXX  

* ### Data that has Sth is named HX.XXX.dat

### The columns in all cases are the same

Some of the original data files have invalid lines or do not satisfy the naming convention, so before doing this I put all the data in the directory called QMCdatacopy and did any modifications to the files there. 

```python
from matplotlib import rc
rc('font', **{'family':'serif'})
rc('text', usetex=True)

import glob 
import os 

# Keep in mind, the data files are 
#     saf files are:   mu  n  n_err  Spi/n  err  Sth/n  err
# entropy files are:   mu  n  n_err  entropy
# density files are:   mu  n  n_err 

# The output files will be (8 columns)
#     mu  n  n_err  Spi/n  err  Sth/n  err  entropy


# The QMC data needs 3 signficant digits in the temperature.
# Here is the correspondance
Tvaldict = {\
    0.29:0.294,  # 20 / 68
    0.35:0.344,  # 20 / 58
    0.36:0.357,  # 20 / 56
    0.37:0.370,\
    0.38:0.385,\
    0.40:0.400,\
    0.50:0.500,\
    0.67:0.667,\
    1.00:1.000,\
    1.111:1.111,\
    0.909:0.909,\
    0.714:0.714,\
    0.667:0.667,\
    0.625:0.625,\
    0.526:0.526,\
    0.476:0.476,\
    0.417:0.417,\
    0.385:0.385,\
    0.357:0.357,\
    0.286:0.286,\
    0.25:0.250,\
    2.4:2.400,\
    4.8:4.800,\
    1.6:1.600,\
    1.11:1.111,\
    1.2:1.200,\
    0.91:0.909,\
    0.71:0.714,\
    0.53:0.526,\
    0.48:0.476,\
    0.42:0.417,\
    0.39:0.385,\
    0.36:0.357,\
    0.30:0.300,\
    0.294:0.294,\
    0.333:0.333,\
    0.323:0.323,\
    0.3125:0.313,\
    0.345:0.345,\
    0.384:0.384,\
    0.478:0.478,\
    0.325:0.325,\
    0.303:0.303,\
    0.32:0.323,\
    
   }


qmcdict_U_SPI = {}
qmcdict_T_SPI = {}

qmcdict_U_STH = {}
qmcdict_T_STH = {}

qmcdict_U_ENTR = {}
qmcdict_T_ENTR = {}

qmcdict_U_DENS = {}
qmcdict_T_DENS = {}

def addto( key, val, dic):
    if key in dic.keys():
        dic[key].append(val)
    else:
        dic[key] = [val]

import pprint 
import collections

def catch_repeated(fs):
    # Check if there are any repeated temperatures 
    Tvals = [] 
    for f in fs:
        Tval0 = float(f.split('T')[1].split('dt')[0])
        Tval = Tvaldict[ Tval0 ]
        Tvals.append(Tval)
    RepeatedTs = [x for x, y in collections.Counter(Tvals).items() if y > 1]
    if len(RepeatedTs)>0:
        print
        print "WARNING: There are repeated temperatures:", RepeatedTs
        pprint.pprint( fs) 
        print   

datadir = '../QMCdatacopy/'
fname = datadir + 'U*'
us = sorted([ float(u.split('/U')[-1]) for u in glob.glob(fname) ] )

for Uval in us: 
    print
    print "U =",Uval
    
    
    # DEALING WITH SAF DATA:
    print "Spi data:"
    fname = datadir + 'U{U:02d}/saf*dat'.format(U=int(Uval))
    fs = sorted(glob.glob(fname))
    catch_repeated(fs)
    
    for i,f in enumerate(fs):       
        dat = None
        for i in range(10):
            try:
                dat = np.loadtxt(f, skiprows=i)
                break
            except:
                if i == 9:
                    print "\n\n problems:"
                    print f, " : cound not load data file after skipping 9 rows"
                    print 
                pass
        
        try:
            Tval0 = float(f.split('T')[1].split('dt')[0])
            Tval = Tvaldict[ Tval0 ]
        except Exception as e:
            print "\n\n problems:"
            print f, " : could not extract Tval" 
            raise e
            
        # Ignore temperature that have chemical potential gaps
        if Uval >= 4. and Tval < 0.30:
            continue
        if Uval >= 8. and Tval < 0.35:
            continue
        
        if Tval == 0.50 and Uval == 14.:
            dat = None
            
        # Some reference data will be used as a comparison to get rid of outlier
        # points in some of the lowset T data 
        refdat = None
        if Uval == 8. and Tval == 0.370:
            try:
                refname = '../QMC_Final/U{:02d}/'.format(int(Uval)) + 'T0.385.dat'
                refdat = np.loadtxt( refname )
            except:
                refdat = None
                
            if len(refdat)==len(dat) and dat is not None and refdat is not None:
                print "\n...Cleaning up U={:0.2f}, T={:0.2f}".format(Uval,Tval)
                spicond   =  np.abs( refdat[:,3] - dat[:,3])  > 0.27
                denscond1 =  np.logical_and( dat[:,1] < 0.92 , dat[:,1] > 0.7)
                denscond2 =  np.logical_and( dat[:,1] > 1.08 , dat[:,1] < 1.3)
                denscond  =  np.logical_or( denscond1, denscond2 )
                l0 =  len( dat[:,3] ) 
                dat = dat[  np.where( ~np.logical_and( spicond, denscond) ) ] 
                l1 =  len( dat[:,3] ) 
                print "...removed {:d} points\n".format( l0-l1 )
                
        if Uval == 9. and Tval == 0.40:
            try:
                refname = '../QMC_Final/U{:02d}/'.format(int(Uval)) + 'T0.500.dat'
                refdat = np.loadtxt( refname )
            except:
                refdat = None
                
            if dat is not None and refdat is not None:
                print "\n...Cleaning up U={:0.2f}, T={:0.2f}".format(Uval,Tval)
                spicond   =  np.abs( refdat[:,3] - dat[:,3])  > 0.25
                denscond1 =  np.logical_and( dat[:,1] < 0.9 , dat[:,1] > 0.77)
                denscond2 =  np.logical_and( dat[:,1] > 1.1 , dat[:,1] < 1.23)
                denscond  =  np.logical_or( denscond1, denscond2 )
                l0 =  len( dat[:,3] ) 
                dat = dat[  np.where( ~np.logical_and( spicond, denscond) ) ] 
                l1 =  len( dat[:,3] ) 
                print "...removed {:d} points\n".format( l0-l1 )
                
                
                
        
        if dat is not None:
            a = np.empty( (dat.shape[0],8) )   
            if dat.shape[1] != 7 :
                print "\n\n problems:"
                print f, " :shape is not 7.  shape is =",dat.shape
                
                
                # Correct some data that has incorrect columns 
                if Uval == 2. and (Tval0==0.667 or Tval0==1.0):
                    print "...correcting data table (shape!=7)"    
                    a[:,0] = dat[:,0]   # chemical potential
                    a[:,1] = dat[:,1]   # density
                    a[:,2] = dat[:,2]   # density err
                    a[:,3] = dat[:,3]   # spi/n
                    a[:,4] = np.zeros_like(dat[:,0])  # spin/n err
                    a[:,5] = dat[:,4]   # sth/n 
                    a[:,6] = np.zeros_like(dat[:,0])  # sth/n err
                    a[:,7] = np.zeros_like(dat[:,0])
                    
                elif Uval == 8. and (Tval0==0.50 or Tval0==0.67):
                    print "...correcting data table (shape!=7)"
                    # These sets are:  mu dens dens_err spi/n
                    a[:,0] = dat[:,0]  # chemical potential
                    a[:,1] = dat[:,1]  # density
                    a[:,2] = dat[:,2]  # density err 
                    a[:,3] = dat[:,3]  # spi/n 
                    a[:,4] = np.zeros_like(dat[:,0])  
                    a[:,5] = np.zeros_like(dat[:,0])  
                    a[:,6] = np.zeros_like(dat[:,0])  
                    a[:,7] = np.zeros_like(dat[:,0])  
                    
                else:
                    print "...ignored error. skipping data set" 
                    a = None
                print 
                
                    
            # For all other data add empty entropy column
            else:
                a = np.empty( (dat.shape[0],8) )   
                for i in range(7):
                    a[:,i] = dat[:,i]
                a[:,7] = np.zeros_like( dat[:,0] )
                
            
                
            if a is not None:
                #print a.shape
                print "T={:0.3f}".format(Tval), 
                # Here go a head and fix the Spi/n for lower densities such that Spi/n = 1 
                a[:,3][  a[:,1] < 0.1 ] = 1.0 
                a[:,5][  a[:,1] < 0.1 ] = 1.0 

                # Move the chemical potential so that n=1 at mu=U/2 
                a[:,0] = a[:,0] + Uval/2. 

                savedir = '../QMC_Final/U{:02d}/'.format(int(Uval))
                if not os.path.exists(savedir):
                    os.makedirs(savedir)

                np.savetxt(savedir + 'T{:.3f}.dat'.format(Tval), \
                   a, fmt='%8.3g' ) 
                
                SthSkip = [(2., 1.0)]
                SthSkip = []
                if (Uval,Tval0) not in SthSkip:
                    np.savetxt(savedir + 'H{:.3f}.dat'.format(Tval), \
                       a, fmt='%8.3g' ) 
                    
                    addto( Uval, Tval, qmcdict_U_STH)
                    addto( Tval, Uval, qmcdict_T_STH)
                else:
                    print
                    print "SKIPPED Stheta.  U={:0.1f}, T={:0.3f}".format(Uval,Tval)
                    
                
                addto( Uval, Tval, qmcdict_U_SPI)
                addto( Tval, Uval, qmcdict_T_SPI)
                
                
    # END OF SAF DATA
    print 
                
            
    # DEALING WITH ENTROPY DATA:
    print "Entropy data:"
    fname = datadir + 'U{U:02d}/entropy*dat'.format(U=int(Uval))
    fs = sorted(glob.glob(fname))
    catch_repeated(fs)
#    if Uval == 8:
#        print fs
    for i,f in enumerate(fs):       
        dat = None
        for i in range(10):
            try:
                dat = np.loadtxt(f, skiprows=i)
                break
            except:
                if i == 9:
                    print "\n\n problems:"
                    print f, " : cound not load data file after skipping 9 rows"
                    print 
                pass
        
        
        Tval0 = float(f.split('T')[1].split('dt')[0])
        Tval = Tvaldict[ Tval0 ]
        
        # Ignore temperature that have chemical potential gaps
        if Uval >= 4. and Tval < 0.30:
            continue
        if Uval >= 8. and Tval < 0.35:
            continue

        if dat is not None:
            a = np.empty( (dat.shape[0],8) )   
            if dat.shape[1] != 4 :
                print f, " :entropy data shape is not 4"
                #print "Uval={:0.2f}, Tval0={:0.2f}".format(Uval,Tval0)
                
                # Correct some data that has incorrect columns 
                if Uval == 12. and Tval0==1.0:
                    print "...correcting data table (shape!=4)"    
                    a[:,0] = dat[:,0]   # chemical potential
                    a[:,1] = dat[:,1]   # density
                    a[:,2] = np.zeros_like(dat[:,0])   # density err
                    for i in [3,4,5,6]:  # Spi/n  err  Sth/n  err
                        a[:,i] = np.zeros_like(dat[:,0])
                    a[:,7] = dat[:,2]
                else:
                    print "...ignored error" 
                    a = None
                print 
                
                
            # For all other data add empty S columns
            else:
                a = np.empty( (dat.shape[0],8) )   
                for i in [0,1,2]:  # mu  n  n_err
                    a[:,i] = dat[:,i]
                a[:,7]= dat[:,3]  # entropy
                for i in [3,4,5,6]:  # Spi/n  err  Sth/n  err
                    a[:,i] = np.zeros_like(dat[:,0])

            
            if a is not None:
                #print a.shape
                print "T={:0.3f}".format(Tval), 
                # Move the chemical potential so that n=1 at mu=U/2 
                a[:,0] = a[:,0] + Uval/2. 

                savedir = '../QMC_Final/U{:02d}/'.format(int(Uval))
                if not os.path.exists(savedir):
                    os.makedirs(savedir)


                np.savetxt(savedir + 'S{:.3f}.dat'.format(Tval), \
                   a, fmt='%8.3g' ) 
                
                addto( Uval, Tval, qmcdict_U_ENTR)
                addto( Tval, Uval, qmcdict_T_ENTR)
    # END OF ENTROPY DATA
    print 

    
            
    # DEALING WITH DENSITY DATA:
    dens_datadir = '../QMCdatacopy/'
    print "Density data:"
    fname = dens_datadir + 'U{U:02d}/dens*dat'.format(U=int(Uval))
    fs = sorted(glob.glob(fname))
    catch_repeated(fs)
#    if Uval == 8:
#        print fs
    for i,f in enumerate(fs):       
        dat = None
        for i in range(10):
            try:
                dat = np.loadtxt(f, skiprows=i)
                break
            except:
                if i == 9:
                    print "\n\n problems:"
                    print f, " : cound not load data file after skipping 9 rows"
                    print 
                pass
        
        
        Tval0 = float(f.split('T')[1].split('dt')[0])
        Tval = Tvaldict[ Tval0 ]
        
        # Ignore temperature that have chemical potential gaps
        if Uval >= 4. and Tval < 0.30:
            continue
        if Uval >= 8. and Tval < 0.35:
            continue
        
        if Uval == 8:
            if Tval in [0.25,0.333]:
                continue

        if dat is not None:
            a = np.empty( (dat.shape[0],8) )   
            if dat.shape[1] != 3 :
                print f, " :density data shape is not 4"
                #print "Uval={:0.2f}, Tval0={:0.2f}".format(Uval,Tval0)
                
                
            # For all other data add empty S columns
            else:
                a = np.empty( (dat.shape[0],8) )   
                for i in [0,1,2]:  # mu  n  n_err
                    a[:,i] = dat[:,i]
                for i in [3,4,5,6,7]:  # Spi/n  err  Sth/n  err
                    a[:,i] = np.zeros_like(dat[:,0])

            
            if a is not None:
                #print a.shape
                print "T={:0.3f}".format(Tval), 
                # Move the chemical potential so that n=1 at mu=U/2 
                a[:,0] = a[:,0] + Uval/2. 

                savedir = '../QMC_Final/U{:02d}/'.format(int(Uval))
                if not os.path.exists(savedir):
                    os.makedirs(savedir)


                np.savetxt(savedir + 'n{:.3f}.dat'.format(Tval), \
                   a, fmt='%8.3g' ) 
                
                addto( Uval, Tval, qmcdict_U_DENS)
                addto( Tval, Uval, qmcdict_T_DENS)
    # END OF DENSITY DATA
    print 
        
import pprint 

print 
print "=================="
print "FINAL DATA SUMMARY"
print "=================="

print "SPI:"
print 
pprint.pprint( qmcdict_U_SPI)
#print 
#pprint.pprint( qmcdict_T_SPI)

print
print "ENTROPY:"
print 
pprint.pprint( qmcdict_U_ENTR)
#print 
#pprint.pprint( qmcdict_T_ENTR)

print
print "DENSITY:"
print 
pprint.pprint( qmcdict_U_DENS)
#print 
#pprint.pprint( qmcdict_T_DENS)

print 
print sorted( np.unique( qmcdict_T_SPI.keys() + qmcdict_T_ENTR.keys()) )
```

## Filtering out some of the noisy $S_{\theta}$ data.  

Only run this cell once after running CELL1.

```python
from savitzky_golay import savitzky_golay

# (Uval,Tval, GOLAYn)
fix = [
  (8., 0.370, 17),\
  (8., 0.385, 17),\
  (8., 0.400, 17),\
  (9., 0.400, 37),\
  (9., 0.500, 11) 
]



# data columns 
MUCOL = 0
DENSCOL = 1
SPInCOL = 3
STHnCOL = 5
ENTRCOL = 7

fig, ax = plt.subplots( 3,2, figsize=(12., 12.) )

axL = [item for sublist in ax for item in sublist]

for i,fx in enumerate(fix):
    dat = np.loadtxt('../QMC_Final/U{:02d}/H{:0.3f}.dat'.format(int(fx[0]), fx[1]) )
    axL[i].plot( dat[:,MUCOL], dat[:,STHnCOL])
    axL[i].set_xlim( -10., 24.)
    axL[i].set_ylim( -0.2, 1.1)
    
    x = dat[:,MUCOL]
    y = dat[:,STHnCOL]
    
    subset = np.logical_and( y>=0., y<=1.05 )
    
    GOLAYWINDOW_Sth = 15
    ySTHhat = np.clip( savitzky_golay( y[subset], GOLAYWINDOW_Sth, 3 ), 0., 1.01)
    
    axL[i].plot( x[subset], ySTHhat)
    
    axL[i].text( 0.95, 0.95, 'U=%d, T=%.3f'%(fx[0], fx[1]),\
                ha='right', va='top',  transform=axL[i].transAxes)
    
    a = np.empty( (len( dat[:,0][subset]), 8) )
    
    print "...creating filtered data "    

    
    a[:,0] = dat[:,0][subset]   # chemical potential
    a[:,1] = dat[:,1][subset]   # density
    a[:,2] = dat[:,2][subset]   # density err
    a[:,3] = dat[:,3][subset]   # spi/n
    a[:,4] = dat[:,4][subset]  # spin/n err
    a[:,5] = ySTHhat   # sth/n 
    a[:,6] = dat[:,6][subset]  # sth/n err
    a[:,7] = np.zeros_like(dat[:,0][subset])  # entropy 
    
    savedir = '../QMC_Final/U{:02d}/'.format(int(fx[0]))
    np.savetxt(savedir + 'H{:.3f}.dat.filter'.format(fx[1]), \
                       a, fmt='%8.3g' ) 
    np.savetxt(savedir + 'H{:.3f}.dat'.format(fx[1]), \
                       a, fmt='%8.3g' ) 
    
    np.savetxt(savedir + 'H{:.3f}.dat.unfiltered'.format(fx[1]), \
                       dat, fmt='%8.3g' ) 
    
    

```

## Final data files have (at most) 8 columns:

    0 chemical potential
    1 density
    2 entropy
    3 spi per particle
    4 compressibility normalized
    5 compressibility bare
    6 sth per particle
    7 sth per site

```python
from kappa import compressibility, compressibility_bare, kappa0

# Save the SPI data
for Uval in qmcdict_U_SPI.keys():
    for Tval in qmcdict_U_SPI[ Uval ]:
        #print 
        #print 'U = {:02d},  T = {:.3f}'.format(int(Uval),Tval)
        fname = '../QMC_Final/U{:02d}/T{:.3f}.dat'.format(int(Uval),Tval)
        dat = np.loadtxt(fname)
        
        # The SPI data will be reduced in the number of 
        a = np.empty( (dat.shape[0], 7))
        a[:,0] = dat[:,0] # chemical potential
        a[:,1] = dat[:,1] # density
        a[:,2] = np.zeros_like( dat[:,0]) # entropy
        a[:,3] = dat[:,3] # spi
        a[:,6] = dat[:,5] # stheta
        
        ndat  = dat[:,1]
        mudat = dat[:,0]
        a[:,4] = compressibility( mudat, ndat )
        a[:,5] = compressibility_bare( mudat, ndat )
        
        if Uval == 8. and Tval==0.5:
            print "Running tests of compressiblity calculation:"
            gspin = 2. 
            numfactor = (6.* np.pi**2.)**(2./3.) / np.pi**2  / gspin**(2./3.)
            print "numfactor = ", numfactor
            plt.plot( ndat, a[:,4] * numfactor, 'r' )
            plt.plot( ndat, a[:,5]/kappa0( ndat, 2.), 'g' )
            
            #print a[:,5]/kappa0(ndat,2.)  / ( a[:,4] * numfactor )
            
            #plt.plot( ndat,  a[:,4] / a[:,5], 'b' )
            #plt.plot( ndat, (2./3.)*(ndat**(5./3.))+0.05, 'k')
            
        
        savedir = '../COMB_Final_Spi/U{:02d}/'.format(int(Uval))
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        fnameout = savedir + 'T{:.3f}.dat'.format(Tval)
        np.savetxt( fnameout, a, fmt='%8.3g')
        #print Uval, Tval, fname 
        
# Save the ENTR data
for Uval in qmcdict_U_ENTR.keys():
    for Tval in qmcdict_U_ENTR[ Uval ]:
        fname = '../QMC_Final/U{:02d}/S{:.3f}.dat'.format(int(Uval),Tval)
        dat = np.loadtxt(fname)
        
        # The SPI data will be reduced in the number of 
        a = np.empty( (dat.shape[0], 4))
        a[:,0] = dat[:,0] # chemical potential
        a[:,1] = dat[:,1] # density
        a[:,2] = dat[:,7] # entropy
        a[:,3] = np.zeros_like( dat[:,0] )
        
        savedir = '../COMB_Final_Entr/U{:02d}/'.format(int(Uval))
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        fnameout = savedir + 'S{:.3f}.dat'.format(Tval)
        np.savetxt( fnameout, a, fmt='%8.3g')
        #print Uval, Tval, fname 
        
        
# Save the DENS data
for Uval in qmcdict_U_DENS.keys():
    for Tval in qmcdict_U_DENS[ Uval ]:
        #print 
        #print 'U = {:02d},  T = {:.3f}'.format(int(Uval),Tval)
        fname = '../QMC_Final/U{:02d}/n{:.3f}.dat'.format(int(Uval),Tval)
        dat = np.loadtxt(fname)
        
        # 
        a = np.empty( (dat.shape[0], 6))
        a[:,0] = dat[:,0] # chemical potential
        a[:,1] = dat[:,1] # density
        a[:,2] = np.zeros_like( dat[:,0]) # entropy
        a[:,3] = np.zeros_like( dat[:,0]) # entropy
        
        ndat  = dat[:,1]
        mudat = dat[:,0]
        a[:,4] = compressibility( mudat, ndat )
        a[:,5] = compressibility_bare( mudat, ndat )
        
        if Uval == 8. and Tval==0.5:
            print "Running tests of compressiblity calculation:"
            gspin = 2. 
            numfactor = (6.* np.pi**2.)**(2./3.) / np.pi**2  / gspin**(2./3.)
            print "numfactor = ", numfactor
            plt.plot( ndat, a[:,4] * numfactor, 'r' )
            plt.plot( ndat, a[:,5]/kappa0( ndat, 2.), 'g' )
            
            #print a[:,5]/kappa0(ndat,2.)  / ( a[:,4] * numfactor )
            
            #plt.plot( ndat,  a[:,4] / a[:,5], 'b' )
            #plt.plot( ndat, (2./3.)*(ndat**(5./3.))+0.05, 'k')
            
        
        savedir = '../COMB_Final_Dens/U{:02d}/'.format(int(Uval))
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        fnameout = savedir + 'n{:.3f}.dat'.format(Tval)
        np.savetxt( fnameout, a, fmt='%8.3g')
        #print Uval, Tval, fname 
        
        
# Save the STH data
for Uval in qmcdict_U_STH.keys():
    for Tval in qmcdict_U_STH[ Uval ]:
        fname = '../QMC_Final/U{:02d}/H{:.3f}.dat'.format(int(Uval),Tval)
        dat = np.loadtxt(fname)
        
        # STH data
        a = np.empty( (dat.shape[0], 8))
        a[:,0] = dat[:,0] # chemical potential
        a[:,1] = dat[:,1] # density
        a[:,2] = np.zeros_like( dat[:,0]) # entropy
        a[:,3] = dat[:,3] # spi
        a[:,4] = np.zeros_like( dat[:,0] )
        a[:,5] = np.zeros_like( dat[:,0] )
        a[:,6] = dat[:,5] # stheta
        a[:,7] = dat[:,5] * dat[:,1] # stheta per lattice site
        
    
        
        savedir = '../COMB_Final_Sth/U{:02d}/'.format(int(Uval))
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        fnameout = savedir + 'H{:.3f}.dat'.format(Tval)
        np.savetxt( fnameout, a, fmt='%8.3g')
        #print Uval, Tval, fname 
```
