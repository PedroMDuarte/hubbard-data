## The data is sorted out properly, but Ehsan's extrapolated data has a lot of noise in it.    

Using a type of low pass filter called Savitzky-Golay filter I was able to smooth out the data and then redefine a data set that makes sense.   The final data set is stored in NLCE7_Final  

The cell below takes care of obtaining and saving this final data set. 

```python
import os

from matplotlib import rc
rc('font', **{'family':'serif'})
rc('text', usetex=True)

import sys
sys.path.append('../')

from savitzky_golay import savitzky_golay

for Tval in [0.40, 0.50, 0.64, 0.72, 0.84, 0.90, 1.0, 1.2, 1.6, 2.0, 2.4, 2.8, 3.2, 3.6, 4.0, 4.6, 5.2, 6.0, 7.0, 8.0, 9.0 ]:
#for Tval in [0.40, 0.50]:

    rows = 9 
    fig = plt.figure(figsize=(15.,11.))
    gsA = matplotlib .gridspec.GridSpec(rows,3, \
        left=0.04, right=0.48, bottom=0.05, top=0.96)
    gsB = matplotlib .gridspec.GridSpec(rows,3, \
        left=0.54, right=0.98, bottom=0.05, top=0.96)
    axL0 = [ fig.add_subplot( gsA[i,0] ) for i in range(rows)] + [ fig.add_subplot( gsB[i,0] ) for i in range(rows)]
    axL1 = [ fig.add_subplot( gsA[i,1] ) for i in range(rows)] + [ fig.add_subplot( gsB[i,1] ) for i in range(rows)]
    axL2 = [ fig.add_subplot( gsA[i,2] ) for i in range(rows)] + [ fig.add_subplot( gsB[i,2] ) for i in range(rows)]

    MUCOL = 0
    STHCOL = 2 

    for i,Uval in enumerate(range(0,31,2)):
    #for i,Uval in enumerate([7]):

        ax0 = axL1[i]
        ax1 = axL2[i]
        ax2 = axL0[i]

        fname =  '../NLCE7_All/U{:02d}/T{:0.2f}.dat'.format(int(Uval),Tval)
        dat = np.loadtxt(fname)
        # plot the unfiltered data
        ax2.plot( dat[:,MUCOL], dat[:,STHCOL], color='k', alpha=0.3)

        

        mu = dat[:,MUCOL]
        ySTH = dat[:,STHCOL]
        sub = np.logical_and( ySTH > -0.1 , ySTH < 1.5) 
        
        
        # compute the filtered out sth
        if Tval < 0.6:
            GOLAYWINDOW_Sth = 31
        elif Tval < 1.3:
            GOLAYWINDOW_Sth = 11
        else:
            GOLAYWINDOW_Sth = 5
        ySTHhat = np.clip( savitzky_golay( ySTH[sub], GOLAYWINDOW_Sth, 3 ), 0., 10.)
        muhat = mu[sub]
        ax2.plot( muhat, ySTHhat ) 
        

        # Get the density corresponding to the chemical potential, not included in Ehsan's dataset  
        
        fname =  '../NLCE8_Final/U{:02d}/T{:0.2f}.dat'.format(int(Uval),Tval)
        dens_dat = np.loadtxt(fname)
        # density data columns
        MUCOLd = 0
        DENSCOLd = 1
        
            
        # Here does an intersection with NLCE8 data to fill up a density column
        if False:
            intersect = np.intersect1d( dens_dat[:,MUCOLd], muhat, assume_unique=True) 
            newdat = np.empty( (len(intersect),4))
            # newdat columns:
            # 0 mu
            # 1 dens
            # 2 sth
            # 3 sth/n


            for ii,muval in enumerate(intersect):
                item = numpy.where( muhat == muval)[0]
                itemdens = numpy.where( dens_dat[:,MUCOLd] == muval )[0]

                try:
                    newdat[ii,0] = muval
                    nval = dens_dat[ itemdens, DENSCOLd]
                    newdat[ii,1] = nval
                    newdat[ii,2] = ySTHhat[item] 
                    if nval < 1e-2:
                        newdat[ii,3] = 1.0 
                    else:
                        newdat[ii,3] = ySTHhat[item]/nval
                except:
                    print "muval = ",  muval
                    print "dens where  = ", itemdens
                    print "muhat where = ", item
                    print 
                    print "shape newdat  = ", newdat.shape
                    print "shape densdat = ", dens_dat.shape
                    print "shape muhat   = ", muhat.shape
                    raise
                    
        # Here it uses only NLCE7 data 
        else:
            newdat = np.empty( (len(muhat),4))
            # newdat columns:
            # 0 mu
            # 1 dens
            # 2 sth
            # 3 sth/n
            newdat[:, 0 ] = muhat
            newdat[:, 1 ] = np.zeros_like(muhat)
            newdat[:, 2 ] = ySTHhat
            newdat[:, 3 ] = np.zeros_like(muhat)
            newdat = newdat[::2]

        
        
        ax0.plot( newdat[:,0], newdat[:,1] )
        ax1.plot( newdat[:,0], newdat[:,3] )
        
        # The new data is saved with a sampling step of 4 to improve speed 
        savedir = '../NLCE7_Final/U{:02d}/'.format(int(Uval))
        if not os.path.exists(savedir):
            os.makedirs(savedir)
        np.savetxt(savedir+'T{:.2f}.dat'.format(Tval), \
                   newdat, fmt='%8.3g') 
        

        #ax0.set_ylim(-0.2, 2.2) 
        ax0.set_ylim(-0.2, 2.2) 
        ax0.set_xlim(-30.,40)

        ax1.set_xlim(-30.,40.)
        ax1.set_ylim(0., 1.4)

        ax2.set_xlim(-30.,40.)
        ax2.set_ylim(-1,1.5)

        ax0.text( 0.05, 0.95, 'U={:02d}'.format(Uval), ha='left', va='top', transform=ax0.transAxes)
    fig.savefig('NLCE7_Final/filtered_all/T{:04.2f}.png'.format(Tval), dpi=300)
    plt.close()
```
